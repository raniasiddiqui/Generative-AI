##**Section 01: Your Solution**

```python
#Add your solution in this block
def calculate(s):
        stack = []
        result = 0
        num = 0
        sign = 1  # 1 represents positive, -1 represents negative

        for char in s:
            if char.isdigit():
                num = num * 10 + int(char)
            elif char == '+':
                result += sign * num
                num = 0
                sign = 1
            elif char == '-':
                result += sign * num
                num = 0
                sign = -1
            elif char == '(':
                stack.append((result, sign))
                result = 0
                sign = 1
            elif char == ')':
                result += sign * num
                num = 0
                prev_result, prev_sign = stack.pop()
                result = prev_result + prev_sign * result

        result += sign * num
        return result
```

```python
print(calculate("1 + 1"))
print(calculate(" 2-1 + 2 "))
print(calculate("(1+(4+5+2)-3)+(6+8)"))
```

## **Section 02: LLM 01 (ChatGPT 4O-mini)**

```python
#Add the code from your chosen LLM here
def calculate(s):
        stack = []
        num = 0
        sign = 1  # 1 means positive, -1 means negative
        result = 0

        for char in s:
            if char.isdigit():
                num = num * 10 + int(char)
            elif char in "+-":
                result += sign * num
                num = 0
                sign = 1 if char == '+' else -1
            elif char == '(':
                stack.append(result)
                stack.append(sign)
                result = 0
                sign = 1
            elif char == ')':
                result += sign * num
                num = 0
                result *= stack.pop()  # Multiply by the sign before '('
                result += stack.pop()  # Add the result before '('

        result += sign * num
        return result
```

## **Section 03: LLM 02 (Gemini)**

```python
#Add the code from your 2nd chosen LLM here
def calculate(s):
        s = s.replace(" ", "")  # Remove spaces
        stack = []
        operand = 0
        sign = 1  # 1 for positive, -1 for negative
        result = 0
        for char in s:
            if char.isdigit():
                operand = operand * 10 + int(char)
            elif char == '+':
                result += sign * operand
                sign = 1
                operand = 0
            elif char == '-':
                result += sign * operand
                sign = -1
                operand = 0
            elif char == '(':
                stack.append(result)
                stack.append(sign)
                sign = 1
                result = 0
            elif char == ')':
                result += sign * operand
                result *= stack.pop()  # Previous sign
                result += stack.pop()  # Previous result
                operand = 0
        result += sign * operand
        return result
```

```python
print(calculate("1 + 1"))
print(calculate(" 2-1 + 2 "))
print(calculate("(1+(4+5+2)-3)+(6+8)"))
```
